```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv('/Users/karinashalya/Downloads/Titanic.csv')
```

```python
titanic
```

```python
import seaborn as sns
```

```python
titanic.drop(['PassengerId', 'Name', 'Ticket', 'Cabin', 'Embarked'], axis=1, inplace=True)
```

```python
mean_age = titanic['Age'].mean()
titanic['Age'].fillna(mean_age, inplace=True)
# Заполнение пропущенных значений в столбце возраста пассажиров
```

```python
titanic
```

```python
plt.figure(figsize=(8, 6))
plt.hist(titanic['Age'], bins=20, edgecolor='k', color='purple')
plt.xlabel('Возраст')
plt.ylabel('Количество пассажиров')
plt.title('Распределение возраста пассажиров')
plt.show()
```

```python
class_counts = titanic['Pclass'].value_counts()
class_counts.plot(kind='bar', figsize=(8, 6),color='red')
plt.xlabel('Класс каюты')
plt.ylabel('Количество пассажиров')
plt.title('Количество пассажиров по классу каюты')
plt.xticks(rotation=0)
plt.show()
```

```python
sns.scatterplot(data=titanic, x='Pclass', y='Fare', color='green')
plt.xlabel('Класс каюты')
plt.ylabel('Стоимость билета')
plt.title('Зависимость стоимости билета от класса каюты')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(titanic['Pclass'], titanic['Age'], alpha=0.5)
plt.xlabel('Класс каюты')
plt.ylabel('Возраст')
plt.title('Отношение возраста и класса каюты')
plt.show()
```

```python
survived_gender = titanic.groupby(['Sex', 'Survived']).size().unstack()
survived_gender.plot(kind='bar', stacked=True, figsize=(8, 6))
plt.xlabel('Пол')
plt.ylabel('Количество')
plt.title('Выжившие и погибшие по полу')
plt.xticks(rotation=0)
plt.legend(['Не выжили', 'Выжили'])
plt.show()
```

```python
class_counts = titanic['Pclass'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Процентное соотношение пассажиров по классу каюты')
plt.axis('equal')
plt.show()
```
